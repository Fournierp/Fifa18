```python
import pandas as pd
import numpy as np

import keras
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras import optimizers
```

```python
model = Sequential([
    Dense(8, input_shape=(3,)),
    Activation('relu'),
    Dense(8),
    Activation('relu'),
    Dense(5),
    Activation('relu'),
    Dense(1),
#     Activation('softmax'),
])
```

```python
pf_optimizer = optimizers.sgd(lr = 0.01)

model.compile(optimizer=pf_optimizer,
              loss='mse', metrics=['mse'])
```

```python
x1 = np.array(np.random.rand(5000))
x2 = np.array(np.random.rand(5000))
x3 = np.array(np.random.rand(5000))
x = np.array([x1,x2,x3]).T

y = x1**2 + 2*x2 + np.sqrt(x3)
```

```python
class PFCallback(keras.callbacks.Callback):
    def __init__(self, test_set, test_labels):
        super().__init__()
        self.test_set = test_set
        self.test_labels = test_labels
        self.previous_epoch_loss = 0

    def on_train_begin(self, logs={}):
        print("Paul Fournier says: training has begun!")
        print("self.params: ", dir(self.params))
        print("self.model: ", dir(self.model))
        print("self.model.optimizer", self.model.optimizer)
        self.epoch_counter = 1
        
    def on_epoch_begin(self, batch, logs={}):
        print("\n\n Paul Fournier says: This is epoch number:" + str(self.epoch_counter) )
        
#     def on_batch_begin(self, batch, logs={}):
#         print(self.model.optimizer.lr)
#         self.model.optimizer.lr -= 0.001
        
    def on_batch_end(self, batch, logs={}):
        pass
        
    def on_epoch_end(self, batch, logs={}):
        #Early stopping 
#         self.losses.append(logs.get('loss'))
        print("Paul Fournier says: " + str(logs) )
        print(np.abs(self.previous_epoch_loss - logs.get('loss')))
        if(np.abs(self.previous_epoch_loss - logs.get('loss')) > 0.0001):
            print("No more learning")
            
        print(logs.get("val_mean_squared_error") - logs.get("mean_squared_error"))
        if(np.abs(logs.get("val_mean_squared_error") - logs.get("mean_squared_error")) > 0.001):
            print("Overfitting")
        self.previous_epoch_loss = logs.get('loss')
        
    def on_train_end(self, logs={}):
        print("\n\n\nTrain ended")
```

```python
pfc = PFCallback(x, y)
h = model.fit(x=x, y=y, epochs=1000, verbose=1, initial_epoch=0, validation_split = 0.2, callbacks=[pfc])
```
